```python
import numpy as np 
import matplotlib.pyplot as plt
import sympy as smp
```

```python
x, y, a, c1, c2, c3, c4 ,v,  b, p, d ,E= smp.symbols('x y α c1 c2 c3 c4 v b p d E')
```

```python
w=smp.sin(a*y)*(1/a**2*(c1*smp.cosh(a*x)+c2*(a*x)*smp.sinh(a*x)+c3*smp.sinh(a*x)+c4*a*x*smp.cosh(a*x)))
w
```

```python
w0=w.subs([(x,0)])
w0
```

```python
C1=smp.solve(w0,c1)
C1
```

```python
dwdx2=smp.diff((smp.diff(w,x)),x)
dwdy2=smp.diff((smp.diff(w,y)),y)
mx=(dwdx2+v*dwdy2)
mx
```

```python
mx0=mx.subs([(x,0),(c1,0)])
mx0
```

```python
C2=smp.solve(mx0,c2)
C2
```

```python
wa=w.subs([(x,10),(c1,0),(c2,0)])
wa
```

```python
ma=mx.subs([(x,10),(c1,0),(c2,0)])-p/d*smp.sin(a*y)
```

```python
sols=smp.solve([wa,ma],[c3,c4])
sols

```

```python
C3=-5*p*a*smp.cosh(10*a)/(d*smp.sinh(10*a)**2)
C3
```

```python
b=10
a=smp.pi/b
p=5
E=25000000
v=0.3
d=E*0.3**3/(12*(1-v**2))
a
```

```python
C4=p/(2*d*smp.sinh(10*a))
C3=-5*p*a*smp.cosh(10*a)/(d*smp.sinh(10*a)**2)
C1=0
C2=0
```

```python
w=smp.sin(a*y)*(1/a**2*(C1*smp.cosh(a*x)+C2*(a*x)*smp.sinh(a*x)+C3*smp.sinh(a*x)+C4*a*x*smp.cosh(a*x)))
```

```python
dwdx2=smp.diff((smp.diff(w,x)),x)
dwdy2=smp.diff((smp.diff(w,y)),y)
mx=(dwdx2+v*dwdy2)
my=(dwdy2+v*dwdx2)
```

```python
omg=smp.lambdify([x,y],w)
Mx=smp.lambdify([x,y],mx)
My=smp.lambdify([x,y],mx)
```

```python
m1 = np.linspace(0,10,100)
m2 = np.linspace(0,10,100)
x1,y1=np.meshgrid(m1,m2)
```

```python
cont=plt.contour(x1,y1,omg(x1,y1),levels=20)
plt.clabel(cont)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16,5))
ax = axes[0]
ax.contour(x1,y1,omg(x1,y1),levels=20)
ax1= axes[1]
ax1.contourf(x1,y1,omg(x1,y1),levels=20, cmap='coolwarm')
```
