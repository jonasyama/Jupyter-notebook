# Lista 6- – Solução geral de cargas a partir de uma placa com carga parcial


```python
import numpy as np 
import matplotlib.pyplot as plt
import sympy as smp
import fourierdouble as fd
```

```python
x,y,a,b,p,wm,v,d,E,f,h,c1,c2,c3,c4, alfa=smp.symbols('x y a b p wm v d E f h c1 c2 c3 c4 alfa')
m,n,i,j=smp.symbols('m n i j', int=True)
```

```python
def derivates(W):
    x,y=smp.symbols('x y')
    dwdx=smp.diff(W,x)
    dwdy=(smp.diff(W,y))
    dwdx2=smp.diff((smp.diff(W,x)),x)
    dwdy2=smp.diff((smp.diff(W,y)),y)
    dwdx3=smp.diff(smp.diff((smp.diff(W,x)),x),x)
    dwdy3=smp.diff(smp.diff((smp.diff(W,y)),y),y)
    dwx2y=smp.diff(smp.diff((smp.diff(W,x)),x),y)
    dwxy2=smp.diff(smp.diff((smp.diff(W,x)),y),y)
    mx=(dwdx2+v*dwdy2)*d
    my=(dwdy2+v*dwdx2)*d
    mxy=(1-v)*(smp.diff((smp.diff(W,x)),y))
    vx=-d*(dwdx3+(2-v)*dwxy2)
    vy=-d*(dwdy3+(2-v)*dwx2y)
    dx=-h/2*dwdx2
    dy=-h/2*dwdy2
    gama=-h*(smp.diff((smp.diff(W,x)),y))
    sigmax=E/(1-v)*(dx+v*dy)
    sigmay=E/(1-v)*(v*dx+dy)
    tal=E/2/(1-v)*gama
    
    return dwdx,dwdy,dwdx2, dwdy2, dwdx3, dwdy3,dwdy3,dwx2y,dwxy2,mx,my,mxy,vx,vy,dx,dy,gama,sigmax,sigmay,tal
```

```python
E=25000000
v=0.2
h=0.3
d=E*h**3/(12*(1-v**2))
a=10
b=10
alfa=n*x*smp.pi/a
```

#### 1) Placa carregamento concentrado


```python
k=fd.sfourier_etaxsi(10,10,10,7,5,0.2,0.2,10,10)
fmn=k[0]
```

```python
def sol_navier(fmn,d,a,b,mm,nn, all_support=True):
    
    x,y,c1,c2,c3,c4=smp.symbols('x y c1 c2 c3 c4')
    m,n=smp.symbols('m n', int=True)
    
    if all_support:
        wmn=fmn/(d*smp.pi**4*((m**2/a**2)+(n**2/b**2))**2)
        freq=smp.sin(x*smp.pi*m/a)*smp.sin(y*smp.pi*n/b)
        W=smp.summation((smp.summation(wmn*freq,(n,1,nn))),(m,1,mm))
        return W
```

```python
w=sol_navier(fmn,d,a,b,10,10)
W=smp.lambdify([x,y],w)
d1=derivates(w)
vx=smp.lambdify([x,y],d1[12])
vy=smp.lambdify([x,y],d1[13])
```

```python
m1 = np.linspace(0.0001,a-0.0001,200)
m2 = np.linspace(0.0001,b-0.0001,200)
x1,y1=np.meshgrid(m1,m2)
```

```python
plt.contourf(x1,y1,W(x1,y1),levels=20, cmap='coolwarm')
plt.colorbar()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16,5))
ax = axes[0]
ax.contourf(x1,y1,vx(x1,y1),levels=10,)
ax1= axes[1]
ax1.contourf(x1,y1,vy(x1,y1),levels=10, cmap='coolwarm')
```

### 2) Carregamento corpo todo

```python
k1=fd.sfourier_etaxsi(10,10,10,5,5,10,10,10,10)
fmn1=k1[0]
```

```python
w1=sol_navier(fmn1,d,a,b,10,10)
W1=smp.lambdify([x,y],w1)
```

```python
plt.contourf(x1,y1,W1(x1,y1),levels=20, cmap='coolwarm')
plt.colorbar()
```
