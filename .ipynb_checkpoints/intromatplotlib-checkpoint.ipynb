```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d
```

```python
x=np.linspace(0,15,30)
y=np.sin(x)+0.1*np.random.randn(len(x))
```

```python
plt.plot(x,y,'o--')
```

```python
plt.figure(figsize=(8,3))
plt.plot(x,y,'o--',label='data1')
plt.xlabel('Time [s] ')
plt.ylabel('Voltage[W]')
plt.legend(loc='lower right',fontsize=15)
```

```python
x2=np.linspace(0,15,100)
y2=np.sin(x2)
plt.figure(figsize=(8,3))
plt.plot(x,y,'o',label='data1')
plt.plot(x2,y2,label='data2')
plt.xlabel('Time [s] ')
plt.ylabel('Voltage[W]')
plt.legend(loc='lower right',fontsize=15,ncol=2)
plt.xlim(right=16)
```

```python
m1 = np.linspace(0,20,100)
m2 = np.linspace(0,15,100)
x,y=np.meshgrid(m1,m2)
z=0.3*np.sin(np.pi*x/20)*np.sin(np.pi*y/15)
```

```python
plt.contourf(x,y,z,levels=20,cmap='plasma')
plt.colorbar(label='w [cm]')
```

```python
c=plt.contour(x,y,z,levels=15)
plt.clabel(c)
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(x,y,-z,cmap='coolwarm',)
ax.set_zlim3d(-1,0)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16,5))
ax = axes[0]
ax.contour(x,y,z)
ax1= axes[1]
ax1.contourf(x,y,z)
```
