```python
%matplotlib widget
import numpy as np 
import matplotlib.pyplot as plt
import sympy as smp
from mpl_toolkits.mplot3d import axes3d, Axes3D
from sympy.plotting import plot3d
import plotly.graph_objects as go
```

```python
x,y,a,b,c=smp.symbols('x y a b c')
```

```python
x1 = np.zeros(100)
y1 = np.linspace(-5, 5, 100)
z1 = np.linspace(-5, 5, 100)
x2 = np.linspace(-5, 5, 100)
y2 = np.linspace(-5, 5, 100)
z2 = np.zeros(100)
x3 = np.linspace(-5, 5, 100)
y3 = np.zeros(100)
z3 =  np.linspace(-5, 5, 100)

plane1 = go.Surface(x=x1, y=y1, z=np.array([z1] * len(x1)))
plane2 = go.Surface(x=x2, y=y2, z=np.array([z2] * len(x1)))
plane3 = go.Surface(x=x3, y=y3, z=np.array([z3] * len(x1)))
```

```python
fig = go.Figure(data=[plane1,plane2])

fig.update_layout(
    autosize=False,
    width=500,
    height=500,scene_aspectmode='cube')
fig.show()
```

```python

x_test_d1 = np.linspace(-4,4,100)
x_test_d2 = np.linspace(-3,3,100)


grid_d1, grid_d2 = np.meshgrid(x_test_d1,x_test_d2)

z = 1.5*np.sin(grid_d1) - 0.1*(grid_d1-3)**2 +10 - 0.5*(grid_d2**2-2) + np.sin(grid_d2)*2


fig = go.Figure(data=[go.Surface(x=grid_d1, y=grid_d2, z=z)])
fig.update_layout(margin=dict(l=20, r=20, t=20, b=20),
    width=800, height=400)

fig.show()
```
