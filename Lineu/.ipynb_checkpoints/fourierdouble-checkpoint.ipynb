# Série dupla de fourier, e cargas concentradas

```python
%matplotlib widget
import numpy as np 
import matplotlib.pyplot as plt
import sympy as smp
from mpl_toolkits.mplot3d import axes3d, Axes3D
from sympy.plotting import plot3d
```

    introdução das bibliotecas e variaveis simbolicas associadas a simplificação de uma função em dupla serie trigonometrica.

```python
x,y,a,b,p,wm,v,d,E,f,h=smp.symbols('x y a b p wm v d E f h')
m,n,i,j=smp.symbols('m n i j', int=True)
```

    definição da função retornavel sfourier_doublesin. É necessario introduzir os parametros (f (função), A e B (dimensoes da placa), nn (numero de funções em x a serem combinadas), mm (numero de funções em y a serem combinadas). A função retorna a plotagem da função, função aproximada pela dupla serie e um valor de erro associado definido pela razão dos volumes sob as superficies.

```python
def sfourier_doublesin(f,a,b,nn,mm):
    freq=smp.sin(x*smp.pi*m/a)*smp.sin(y*smp.pi*n/b)
    fmn=4/a/b*smp.integrate(smp.integrate(f*freq,(x,0,a)),(y,0,b))
    F=smp.summation((smp.summation(fmn*freq,(n,1,nn))),(m,1,mm))
    plot3d(F, (x, 0,a), (y, 0,b))
    plot3d(f, (x, 0,a), (y, 0,b))
   
    V1=smp.integrate(smp.integrate(F,(x,0,a)),(y,0,b))
    V2=smp.integrate(smp.integrate(f,(x,0,a)),(y,0,b))
    erro=V1/V2
    erro.evalf()
    print(erro.evalf())
    return fmn,F
```

```python

def sfourier_value(f,a,b,nn,mm):
    freq=smp.sin(x*smp.pi*m/a)*smp.sin(y*smp.pi*n/b)
    fmn=4/a/b*smp.integrate(smp.integrate(f*freq,(x,0,a)),(y,0,b))
    F=smp.summation((smp.summation(fmn*freq,(n,1,nn))),(m,1,mm))
    return fmn,F
```

    exemplo da função para a função (x^2 + y^2)

```python
eq= sfourier_doublesin(x**2+y**2,10,10,10,10)
```

    definição da função retornavel sfourier_etaxsi. é necessario a introção dos parametros f (função), a e b (dimensões do dominio),xsi (dimensão em x da aplicação do carregamento), eta (dimensão em y da aplicação do carregamento, c (dimensão discreta da parede ou pilar em x), d (dimensão discreta da parede ou pilar em y), n (numero de funções em x a serem combinadas), mm (numero de funções em y a serem combinadas). A função retorna a plotagem função aproximada pela dupla serie

```python
def sfourier_etaxsi(f,a,b,xsi,eta,c,d,nn,mm):
    freq=smp.sin(x*smp.pi*m/a)*smp.sin(y*smp.pi*n/b)
    fmn=4/a/b*smp.integrate(smp.integrate(f*freq,(x,xsi-c/2,xsi+c/2)),(y,eta-d/2,eta+d/2))
    F=smp.summation((smp.summation(fmn*freq,(n,1,nn))),(m,1,mm))
    plot3d(F, (x, 0,a), (y, 0,b))

   


    return fmn,F
```
