```python
import numpy as np 
import matplotlib.pyplot as plt
import sympy as smp
```

```python
b,h,l,e,e2,m,p,x,E,I,c1,c2=smp.symbols('b h l e e2 m p x E I c1 c2')
```

para a caso1


```python
e=0.3*h
```

```python
e=0.3-0.6/(l/2)*x
e2=-0.3+0.3*(x-l/2)/(l/2)
m1=-e*p
m2=-e2*p
```

```python
w=smp.integrate(smp.integrate(0.3*h*p/E/I,(x,0,x))+c1,(x,0,x))
W=w+c2
WW=W.subs([(c1,0),(c2,0)])
WW
```

```python
w0=W.subs([(x,0)])
dwdx=smp.diff(W,x)
dwdx0=dwdx.subs([(x,0)])
sol=smp.solve([w0,dwdx0],[c1,c2])
sol
```

```python
WW=smp.lambdify([h,p,E,I,x],WW)
```

```python
x1=np.linspace(0,10,100)
plt.plot(x1,-WW(0.5,1000,25000000,0.003125,x1))
max=np.min(-WW(0.5,1000,25000000,0.003125,x1))
max
#momento constante
```

```python
m1=p*l-p*x
w1=1/E/I*smp.integrate(smp.integrate(m1,(x,0,x)),(x,0,x))
w1.simplify()
W1=smp.lambdify([p,E,I,l,x],w1)
```

```python
plt.plot(x1,W1(1,25000000,0.003125,10,x1))
max1=np.max(W1(1,25000000,0.003125,10,x1))
max1
```
